```python
def categorize_sentiment(score, rating):
    if score > 0.5:
        if rating >= 4:
            return 'Positive'
        elif rating == 3:
            return 'Mixed Positive' 
        else:
            return 'Mixed Negative'
    elif score < -0.05:
        if rating <= 2:
            return 'Negative'
        elif rating == 3:
            return 'Mixed Positive'
    else:
        if rating >= 4:
            return 'Positive'
        elif rating <= 2:
            return 'Negative'  # Low rating with neutral sentiment
        else:
            return 'Neutral'
```

```python
# Define a function to bucket sentiment scores into text ranges
def sentiment_bucket(score):
    if score >= 0.5:
        return '0.5 to 1.0'  # Strongly positive sentiment
    elif 0.0 <= score < 0.5:
        return '0.0 to 0.49'  # Mildly positive sentiment
    elif -0.5 <= score < 0.0:
        return '-0.49 to 0.0'  # Mildly negative sentiment
    else:
        return '-1.0 to -0.5'  # Strongly negative sentiment

# Apply sentiment analysis to calculate sentiment scores for each review
customer_reviews_df['SentimentScore'] = customer_reviews_df['ReviewText'].apply(calculate_sentiment)

# Apply sentiment categorization using both text and rating
customer_reviews_df['SentimentCategory'] = customer_reviews_df.apply(
    lambda row: categorize_sentiment(row['SentimentScore'], row['Rating']), axis=1)

# Apply sentiment bucketing to categorize scores into defined ranges
customer_reviews_df['SentimentBucket'] = customer_reviews_df['SentimentScore'].apply(sentiment_bucket)

# Display the first few rows of the DataFrame with sentiment scores, categories, and buckets
print(customer_reviews_df.head())

# Save the DataFrame with sentiment scores, categories, and buckets to a new CSV file
customer_reviews_df.to_csv('fact_customer_reviews_with_sentiment.csv', index=False)
```

```python
import pandas as pd

query = """
SELECT
    ReviewID,
    CustomerID,
    ProductID,
    ReviewDate,
    Rating,
    ReviewText
FROM dbo.customer_reviews
"""

customer_reviews_df = pd.read_sql(query, conn)
customer_reviews_df.head()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()

# 1. Function: calculate sentiment score from text
def calculate_sentiment(text):
    if not isinstance(text, str):
        return 0.0
    scores = sia.polarity_scores(text)
    return scores['compound']          # MUST return a number between -1 and 1

# 2. Function: combine rating + sentiment into a category
def categorize_sentiment(score, rating):
    # you can adjust these rules; this is just an example
    if score >= 0.3 and rating >= 4:
        return 'Positive'
    elif score <= -0.3 and rating <= 2:
        return 'Negative'
    elif -0.3 < score < 0.3:
        return 'Neutral'
    else:
        return 'Mixed'

# 3. Function: bucket sentiment score into ranges
def sentiment_bucket(score):
    if score >= 0.5:
        return '0.5 to 1.0'        # Strongly positive
    elif 0.0 <= score < 0.5:
        return '0.0 to 0.49'       # Mildly positive
    elif -0.5 <= score < 0.0:
        return '-0.49 to 0.0'      # Mildly negative
    else:
        return '-1.0 to -0.5'      # Strongly negative

# 4. Apply to your DataFrame (customer_reviews_df must already be loaded)
customer_reviews_df['SentimentScore'] = customer_reviews_df['ReviewText'].apply(calculate_sentiment)

customer_reviews_df['SentimentCategory'] = customer_reviews_df.apply(
    lambda row: categorize_sentiment(row['SentimentScore'], row['Rating']),
    axis=1
)

customer_reviews_df['SentimentBucket'] = customer_reviews_df['SentimentScore'].apply(sentiment_bucket)

print(customer_reviews_df.head())

customer_reviews_df.to_csv('fact_customer_reviews_with_sentiment.csv', index=False)
```
